```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\PC\OneDrive - University of South Africa\Desktop\Mbulelo Majola\Excel\Books_Data_Clean.csv")
df.head(10)
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns = df.columns.str.replace('-','_')
```

```python
df.head(10)
```

```python
df['publishing_year']= pd.to_numeric(df['publishing_year'], errors = 'coerce')
```

```python
NaN_rows = df[df['publishing_year'].isna()]
print(NaN_rows)
```

```python
avg = df['publishing_year'].mean()
df['publishing_year'].fillna(avg, inplace = True)
```

```python
df['book_name'] = df['book_name'].str.replace('\d+', '')
```

```python
mod = df['book_name'].mode()[0]
df['book_name'].fillna(mod, inplace = True)
```

```python
df['language_code'] = df['language_code'].str.replace('\d+', '')
```

```python
mod = df['language_code'].mode()[0]
df['language_code'].fillna(mod, inplace = True)
```

```python
df.dropna(subset = 'book_name', inplace = True)
```

```python
df.duplicated().any()
```

```python
df = df[df['publishing_year'] >1900]
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
plt.hist(df['publishing_year'])
plt.title('Distribution of Book Publishing Years', fontsize=16)
plt.xlabel('Publishing Year')
plt.ylabel('Frequency (Number of Books)', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
df['genre'].value_counts().plot(kind = 'bar')
plt.xlabel('genre')
plt.ylabel('number of books')
plt.show() 
```

```python
df.groupby('author')['book_average_rating'].mean()
```

```python
sb.boxplot(x= 'genre', y= 'book_ratings_count', data = df, color='m')
plt.title('Book_ratings_count of each genre')
plt.xlabel('Genre')
plt.ylabel('Book ratings count')
plt.show()
```

```python
sb.scatterplot(x='sale_price', y='units_sold', data=df, color= 'red')
plt.title('Relationship between Sale Price and Units Sold')
plt.xlabel('Sale Price')
plt.ylabel('Units Sold')
plt.show()
```

```python
language_count = df['language_code'].value_counts()
```

```python
plt.figure(figsize=(10, 8))  
plt.pie(language_count, labels= language_count.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Books by Language')
plt.axis('equal')
plt.show()
```

```python
df.groupby('publisher')['publisher_revenue'].sum().sort_values(ascending = False)
```

```python
df.groupby('author_rating')['book_ratings_count'].mean().sort_values(ascending = False).max()
```

```python
author_sales = df.groupby('author')['gross_sales'].sum().sort_values(ascending=False)
# Create a bar chart to visualize total gross sales by author
plt.figure(figsize=(12, 8))  # Set figure size for better readability
author_sales.head(20).plot(kind='bar', color='green')
# Add a title to the plot
plt.title('Total Gross Sales by Author')
# Label the x-axis
plt.xlabel('Author')
# Label the y-axis
plt.ylabel('Total Gross Sales')
# Adjust layout to prevent label cutoff
plt.tight_layout()
# Display the plot
plt.show()
```
